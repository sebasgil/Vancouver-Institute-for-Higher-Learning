# Classifying Costs 

## Review

Recall that last time we established a few useful criteria used to classify costs. As we continue our studies of how companies work, it will be important to keep in mind how allocating costs in different ways affects the decision making process.

* **fixed cost**: is a cost of operation that will not be affected by output.
* **variable cost**: is a cost whose value depends on the level of production
* **direct cost**: is a kind of cost that is *directly* associated to *each* unit produced
* **indirect cost**: is a cost that cannot be directly associated with each unit produced; also known as **overhead**

## Why is it important to distinguish between direct and indirect cost? 

![buggy](http://1.bp.blogspot.com/-DtMJMbjeeTo/T1zgrPPZ-VI/AAAAAAAA6l0/32nYWbDIYxc/s640/Evolution+of+Volkswagen+Ads+From+1960+%288%29.jpg)

Direct and indirect costs play a [key role](http://www.businessnewsdaily.com/5498-direct-costs-indirect-costs.html) in assigning a price to a product. For the continued operation of a business, it is necessary to charge a price that covers more than just the cost of materials and equipment required to manufacture its products or produce its services. Worker salaries, marketing campaigns, rent and maintenance of the premises are all factors that have to be properly accounted for. 

The majority of direct costs correspond to the labor and direct materials required for a specific product. For example, to create its product, an appliance maker requires steel, electronic components and other raw materials. Direct costs can be either fixed or variable. Worker salaries tend to be a fixed cost since an employee's salary usually remains fixed throughout the year. Other direct costs, like materials, are variable costs because expenses increase as additional units of a product or service are created.

Indirect costs are usually referred to as the "real cost of running business" because they go beyond the costs associated with creating a particular product to include the price of maintaining the entire company. Just like cleaning supplies and maintenance are examples of materials classified as indirect costs, overhead can also take the form of indirect labor. Indirect labor costs make the production of cost objects possible, but aren't assigned to a specific product. For example, clerical assistants who help maintain an office support their parent company as a whole irrespective of any one product line. Thus, their labor can be counted as an indirect cost. Other common indirect costs include advertising and marketing, communication, "fringe benefits" such as an employee gym, and accounting and payroll services.

## Taxation of Costs

![CRA](http://www.hyperactive.ca/uploads/projects/media/images/feature/68384a24af0ea572eb1e2c912c525415e5945a92.jpg)

It is important to correctly classify direct and indirect costs because these are taxed differently. For instance, overhead expenses such as utilities needed to power equipment and the inventory needed to manage an office space are often times tax-deductible. In some cases, even the costs of goods sold qualify for deductions. Thus, negligent classification of costs can result in a company being audited by the revenue agency.

This policy extends beyond the private sector to include public enterprises as well. Government grants such as the NSERC grants required for scientific research usually have very stringent policies for expenses that qualify for direct versus indirect costs, and they allocate specific amounts of funding for a variety of activities such as ordering new equipment, hiring researchers and even the travel costs associated with going to conferences.

# Price and Revenue 

Last time we introduced the **price or demand function** as a decreasing function of $x$. The product of $x$ and $p(x)$ defines the **revenue**. That is to say, if $x$ units are sold at a price of $p(x)$ per unit, the total revenue accrued by the firm will be given by

$$
R(x) = xp(x)
$$

Taking the derivative of this relation yields the **marginal revenue function**, which measures the change in revenue with respect to the number of units sold. Now, if $x$ units are sold, the total **profit** is 

$$
P(x) = R(x) - C(x)
$$

## Example: Freedom Mobile 

![freedom mobile](https://pbs.twimg.com/profile_images/800827353618350080/Rp2pVM8M.jpg)

### Statement of the Problem 

Freedom Mobile, a new mobility company in Vancouver, is currently selling a smartphone and mobility plan bundle for the Samsung Galaxy S8. After signing a two-year contract, costumers will pay the full price of the \$800 phone in monthly installments added to their phone fees. With the new S9 in the works and a desire to sell the remaining phones in stock as soon as possible, a marketing survey reports that for each \$10 rebate offered to buyers, the number of units sold will increase by 20 a week. If 200 smartphone bundles are currently being sold per week, find the price and revenue functions. How large a rebate should Freedom Mobile offer its customers to maximize their revenue?

### Solution

If $x$ is the number of smartphone bundles sold per week, then the weekly increase in sales is given by $x - 200$. For each increase of 20 units sold, the price is decreased by \$10. Thus, for each additional unit sold, the decrease in price will be $\frac{1}{20} \times 10$ and the demand function is:

$$
p(x) = 800 - \frac{10}{20}(x - 200) = 900 - \frac{1}{2}x
$$

```python
import numpy as np
import pylab as plt

x = np.linspace(0,1800,100)
price = 900 - 0.5*x
plt.title('Price Function for Smartphone Bundles')
plt.xlabel('Number of Units Sold')
plt.ylabel('Price of Unit ($)')
plt.plot(x,price, 'r')
plt.show()
```

while the revenue function is

$$
R(x) = xp(x) = 900x - \frac{1}{2}x^2.
$$

```python
revenue = 900*x - 0.5*x**2
plt.title('Revenue Function for Smartphone Bundles')
plt.xlabel('Number of Units Sold')
plt.ylabel('Revenue Accrued ($)')
plt.plot(x,revenue, 'g')
plt.show()
```

We proceed by finding the critical points, $R'(x) = 900 - x$ implies that $R'(x) = 0$ when $x = 900$. Since the graph of the revenue function is a downward facing parabola, we find that $x = 900$ is the critical point that maximizes the revenue. We can now proceed to plug in this value into the price function

$$
p(900) = 800 - \frac{10}{20}(x - 200) = 450
$$

Finally, the rebate is $\Delta p = 800 - 450 = 350$. This means that buyers who purchase the bundle during the rebate period will have to pay \$15 monthly during the duration of their contract.
